```python
#here we are going to convert dictonary to list and removing whole id and only name_id from genres
```

```python
#we used ast to convert string to list as we see the output is string so if we need only name we need to convert it into list
import ast 
def convert3(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(convert3)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert3)
```

```python
movies.head()
```

```python
def convert4(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i['name'])
            counter=counter+1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(convert4)
```

```python
movies
```

```python
    def fetch_director(text):
        L = []
        for i in ast.literal_eval(text):
            if i['job'] == 'Director':
                L.append(i['name'])
        return L 
```

```python
  def collapse(text):
        L1= []
        for i in text:
                L1.append(i.replace(" ",""))
        return L1 
```

```python
movies['keywords']=movies['keywords'].apply(collapse)
movies['cast']=movies['cast'].apply(collapse)
movies['crew']=movies['crew'].apply(collapse)
movies['genres']=movies['genres'].apply(collapse)
```

```python
movies.head()
```

```python
movies['tag']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new_df.head().tag[0]
```

```python
new_df['tag']=new_df['tag'].apply(lambda x : " ".join(x))
```

```python
new_df['tag']=new_df['tag'].apply(lambda x : x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()

def stem(text):
    s=[]
    for i in text.split(): 
        s.append(ps.stem(i))
    return " ".join(s)  

new_df['tag']= new_df['tag'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
cv = CountVectorizer(max_features=5000,stop_words='english')
vector = cv.fit_transform(new_df['tag']).toarray()


```

```python
cv.get_feature_names_out()[0:200]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
cv.get_feature_names_out()[0:100]
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
